```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
n_rows = 1000000
```

```python
latitude = np.random.uniform(low=-90, high=90, size=n_rows)
longitude = np.random.uniform(low=-180, high=180, size=n_rows)
elevation = np.random.uniform(low=-500, high=9000, size=n_rows)
df = pd.DataFrame({
    'latitude': latitude,
    'longitude': longitude,
    'elevation': elevation
})
```

```python
china_df = df[
    (df['latitude'].between(18, 53)) & 
    (df['longitude'].between(73, 135))
]
print(f"Number of points in China: {len(china_df)}")
print("\nFirst few rows of points in China:")
print(china_df.head())
```

```python
portugal_df = df[
    (df['latitude'].between(36.5, 42.2)) & 
    (df['longitude'].between(-9.5, -6))
]
print(f"Number of points in Portugal: {len(portugal_df)}")
print("\nFirst few rows of points in Portugal:")
print(portugal_df.head())
```

```python
country_df = china_df
```
